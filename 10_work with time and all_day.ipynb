```python
def status_of_day(h):
    compare_userloclist_new['third_mode'] = "dsgrtr"
    for i in range(len(compare_userloclist_new)) :
        if compare_userloclist_new['hour'][i] <= 3 and compare_userloclist_new['hour'][i] >= 0 :
            compare_userloclist_new['third_mode'][i] = "time_0-3"
        if compare_userloclist_new['hour'][i] <= 7 and compare_userloclist_new['hour'][i] >= 4 :
            compare_userloclist_new['third_mode'][i] = "time_4-7"
        if compare_userloclist_new['hour'][i] <= 11 and compare_userloclist_new['hour'][i] >= 8 :
            compare_userloclist_new['third_mode'][i] = "time_8-11"
        if compare_userloclist_new['hour'][i] <= 15 and compare_userloclist_new['hour'][i] >= 12 :
            compare_userloclist_new['third_mode'][i] = "time_12-15"   
        if compare_userloclist_new['hour'][i] <= 19 and compare_userloclist_new['hour'][i] >= 16 :
            compare_userloclist_new['third_mode'][i] = "time_16-19"
        if compare_userloclist_new['hour'][i] <= 23 and compare_userloclist_new['hour'][i] >= 20 :
            compare_userloclist_new['third_mode'][i] = "time_20-23"

            
    return compare_userloclist_new
```
