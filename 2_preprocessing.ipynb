```python
dataset_final['check-in time'] = dataset_final['check-in time'].map(lambda x: datetime.strptime(x, "%Y-%m-%d %H:%M:%S"))
dataset_final['checkin_date'] = pd.to_datetime(dataset_final['check-in time']).dt.date
dataset_final['checkin_time'] = pd.to_datetime(dataset_final['check-in time']).dt.strftime('%I:%M %p')
```

```python
dataset_final["DayOfWeek"] = dataset_final["check-in time"].dt.weekday
```

```python
dataset_final["IsWeekend"] = dataset_final["DayOfWeek"] >= 5
```

```python
dataset_final["IsWeekend_int"] = dataset_final["IsWeekend"].astype(int)
dataset_final['IsWeekend_int']
```

```python
dataset_final
```

```python
time_AM_PM = (dataset_final['checkin_time']  == 'PM').astype(int)
dataset_final['time_AM_PM'] = time_AM_PM
```

```python
substring = "PM"
for i , row in dataset_final.iterrows():
    
    if substring in dataset_final['checkin_time'][i] :
        dataset_final['time_AM_PM'][i] = 1
        
    else :
        dataset_final['time_AM_PM'][i] = 0
```

```python
dataset_final
```

```python
AM_PM = dataset_final['time_AM_PM'].map({1: 'PM' , 0: 'AM' })
dataset_final['AM_PM'] = AM_PM
```
