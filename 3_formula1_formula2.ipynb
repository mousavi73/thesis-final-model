```python
dataset_final = pd.read_csv('D://A_thesis/gowalla/dataset_final_with_date_day_ampm.csv')
```

```python
dataset_final
```

```python
dataset_final['checkin_date'] = pd.to_datetime(dataset_final['check-in time']).dt.date
dataset_final['checkin_time'] = pd.to_datetime(dataset_final['check-in time']).dt.strftime('%I:%M %p')
```

```python
dataset_final = dataset_final.sort_values(by=['userid','checkin_date'],ascending = [True, True])
dataset_final
```

```python
max_date = dataset_final.groupby(['userid'])['checkin_date'].max()
min_date = dataset_final.groupby(['userid'])['checkin_date'].min()
```

```python
lifetime_user = pd.merge(max_date, min_date, on='userid')
```

```python
lifetime_user.rename(columns={"checkin_date_x": "last time", "checkin_date_y": "first time"}, inplace = True)
lifetime_user.reset_index(inplace=True)
```

```python
lifetime_user
```

```python
lifetime = lifetime_user['last time'] - lifetime_user['first time']
# lifetime
lifetime_user['lifetime'] = lifetime
```

```python
lifetime_user
```

```python
lifetime_user.to_csv('D://A_thesis/gowalla/out_random/lifetime_user.csv')
```

```python
last_visit = dataset_final.groupby(['userid', 'location_id'])['checkin_date'].max()
# last_visit
first_visit = dataset_final.groupby(['userid', 'location_id'])['checkin_date'].min()
# first_visit
```

```python
merge_visit = pd.merge(last_visit, first_visit, on=["userid","location_id"])
merge_visit
```

```python
merge_visit.rename(columns={"checkin_date_x": "last_visit_LinU", "checkin_date_y": "first_visit_LinU"}, inplace = True)
merge_visit.reset_index(inplace=True)
```

```python
merge_visit
```

```python
difference_visit = merge_visit['last_visit_LinU'] - merge_visit['first_visit_LinU']
merge_visit['difference_visit'] = difference_visit
```

```python
merge_visit
```

```python
merge_visit.to_csv('D://A_thesis/gowalla/out_random/merge_visit.csv')
```
