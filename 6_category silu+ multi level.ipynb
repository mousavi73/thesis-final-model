```python
def category(category_name) :
    global sorted_category_silu_final
    df = pd.DataFrame()
     
    user_id = list(sorted_category_silu_final.userid.unique())
    
    for i in range(len(user_id)):
        df=df.append(sorted_category_silu_final[(sorted_category_silu_final['userid'] == user_id[i]) & (sorted_category_silu_final['category'] == category_name) ])
        

        
    return df
```

```python
for i in range (len(list_2000)):
    df = pd.DataFrame()
    df = category(list_2000[i])
    A_2000[list_2000[i],'visit_user_from_cat'] = list(df['visit_user_from_cat'])
    A_2000[list_2000[i],'visit_user_cat_weekday'] = list(df['visit_user_cat_weekday'])
    A_2000[list_2000[i],'visit_user_cat_weekend'] = list(df['visit_user_cat_weekend'])
    A_2000[list_2000[i],'visit_user_cat_AM'] = list(df['visit_user_cat_AM'])
    A_2000[list_2000[i],'visit_user_cat_PM'] = list(df['visit_user_cat_PM'])
```
